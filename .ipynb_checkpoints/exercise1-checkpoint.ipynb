```python
import pandas as pd # for data manipulation
import matplotlib.pyplot as plt # for plotting
import numpy as np #for numeric calculations and making simulated data.
import seaborn as sns # for plotting, an extension on matplotlib
import sklearn
```

# Task 1

```python
data1 = pd.read_csv("ecg_ppg_clean.csv")
```

```python
data2 = pd.read_csv("ecg_ppg_motion.csv")
```

# clean datasets from NaN values

```python
ppg_cols = [
    'timestamps_ppg_1_green', 'ppg_1_green',
    'timestamps_ppg_1_red',   'ppg_1_red',
    'timestamps_ppg_1_ir',    'ppg_1_ir',
    'timestamps_ppg_2_green', 'ppg_2_green',
    'timestamps_ppg_2_red',   'ppg_2_red',
    'timestamps_ppg_2_ir',    'ppg_2_ir'
]

ecg_cols = ['timestamps_ecg', 'ecg']
clean_ecg = data1[ecg_cols]
motion_ecg = data2[ecg_cols]

clean_ppg = data1[ppg_cols].dropna()
motion_ppg = data2[ppg_cols].dropna()
```

```python
motion_ecg.size
```

```python
def minmax_normalize(df):
    return (df - df.min()) / (df.max() - df.min())
```

```python
norm_clean_ppg  = minmax_normalize(clean_ppg[ppg_cols]) # contains ppg 1 & 2
norm_motion_ppg = minmax_normalize(motion_ppg[ppg_cols]) # contains ppg 1 & 2
```

```python
norm_clean_ecg = clean_ecg.copy()
norm_clean_ecg['ecg'] = minmax_normalize(clean_ecg['ecg'])

norm_motion_ecg = motion_ecg.copy()
norm_motion_ecg['ecg'] = minmax_normalize(motion_ecg['ecg'])
```

```python
norm_clean_ecg.head()
```

```python
ecg_palette = {
    'clean': 'tab:blue',
    'motion': 'tab:orange'
}
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12,4)) # for motion and clean

fig.subplots_adjust(hspace=0.8)

i0, i1 = 4000, 5000

ecg_clean_ds  = norm_clean_ecg.iloc[i0:i1]
ecg_motion_ds = norm_motion_ecg.iloc[i0:i1]

sns.lineplot(
    ax=axes[0],
    x=ecg_clean_ds['timestamps_ecg'],
    y=ecg_clean_ds['ecg'],
    color='tab:blue',
    linewidth=1
)
axes[0].set_title("Clean ECG")
axes[0].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[1],
    x=ecg_motion_ds['timestamps_ecg'],
    y=ecg_motion_ds['ecg'],
    color='tab:orange',
    linewidth=1
)
axes[1].set_title("Motion ECG")
axes[1].set_xlabel("Time (s)")
```

```python
ppg_palette = {
    'ppg_1_green': 'green',
    'ppg_1_red': 'red',
    'ppg_1_ir': 'purple',
    'ppg_2_green': 'green',
    'ppg_2_red': 'red',
    'ppg_2_ir': 'purple'
}
```

# Task 2

Resample both signals so that they are the same length and samples are equidistant. Sample
frequency should be easily adjusted as a variable. Resample the signals to have 200 Hz as the
sampling frequency. The original sampling frequency for ecg is 128 Hz and for ppg it is 100 Hz.
Length of the measurements is just over 2 mins.

```python
fs_ecg = 128
fs_ppg = 100
fs_target = 200 # sample frequency


duration_sec_ecg = len(norm_clean_ecg['ecg']) / fs_ecg
duration_sec_ppg = len(norm_clean_ppg['ppg_1_red']) / fs_ppg

duration_sec = min(duration_sec_ppg, duration_sec_ecg)
print(duration_sec) # should be around 120 sec

n_target = int(duration_sec * fs_target)
print(n_target)
```

```python
from scipy.signal import resample
```

```python
def resample_ppg(ppg):

    return resample(ppg, n_target)

```

```python
ecg_200_clean = resample(norm_clean_ecg['ecg'], n_target)
ecg_200_motion = resample(norm_motion_ecg['ecg'], n_target)


# ppg1 clean full

ppg_1_green_200_clean = resample_ppg(norm_clean_ppg['ppg_1_green'])
ppg_1_ir_200_clean = resample_ppg(norm_clean_ppg['ppg_1_ir'])
ppg_1_red_200_clean = resample_ppg(norm_clean_ppg['ppg_1_red'])

# ppg2 clean full

ppg_2_green_200_clean = resample_ppg(norm_clean_ppg['ppg_2_green'])
ppg_2_ir_200_clean = resample_ppg(norm_clean_ppg['ppg_2_ir'])
ppg_2_red_200_clean = resample_ppg(norm_clean_ppg['ppg_2_red'])

# ppg1 motion full

ppg_1_green_200_motion = resample_ppg(norm_motion_ppg['ppg_1_green'])
ppg_1_ir_200_motion = resample_ppg(norm_motion_ppg['ppg_1_ir'])
ppg_1_red_200_motion = resample_ppg(norm_motion_ppg['ppg_1_red'])

# ppg2 motion full

ppg_2_green_200_motion = resample_ppg(norm_motion_ppg['ppg_2_green'])
ppg_2_ir_200_motion = resample_ppg(norm_motion_ppg['ppg_2_ir'])
ppg_2_red_200_motion = resample_ppg(norm_motion_ppg['ppg_2_red'])

t_200 = np.arange(n_target) / fs_target
```

```python
t0 = 30 # start of slice
window = 10 # slice window

i0 = int(t0 * fs_target)
i1 = i0 + int(window * fs_target) # slice indexes
```

# Task 3

Visualize the signals in frequency domain. Do they look as expected (compare to frequencies
that you expected to see in these modalities)? Which frequency band carries most of the
information?

First analyzing the electrocardiogram (ecg) with fourier transfrom.

Expected result from ecg is the highest peak at around 1-2hz (heart rate is around 60 to 120 (bpm) resting) because heart rate is the most re-occurring event in the ecg signal

```python
from scipy.fft import fft, fftfreq
from scipy.signal import detrend
```

```python
x = ecg_200_clean
N = len(x)

x = x - np.mean(x)

T = 1 / fs_target

X = fft(x)

f = fftfreq(N, T)  # frequencies

fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.lineplot(ax=axes[0], x=f[:N//2], y=np.abs(X[:N//2]), linewidth=(0.8))

axes[0].set_title("ECG_clean frequency spectrum 0-100HZ (fft)")
axes[0].set_xlabel("Frequency (HZ)")

sns.lineplot(ax=axes[1], x=f[:N//8], y=np.abs(X[:N//8]), linewidth=(0.8)) # zoom for better analysis

axes[1].set_title("ECG_clean frequency spectrum 0-30HZ (fft)")
axes[1].set_xlabel("Frequency (HZ)")
```

The most dominant frequency seems to indeed be at around 1 - 1.5HZ (corresponding to my assumption about heart rate). The following peaks are decreasing in amplitude as the frequency increases but I think it is important to note that they do follow a pattern of being integer multiples of the most dominant frequency (1 - 1.5HZ) with the next one being at around 2.3 - 2.7 HZ (meaing 2x) and so on. These are only visible in the clean dataset because the subject is not currently moving and thus adding noise caused by motion. 

```python
x = ecg_200_motion
N = len(x)

x = x - np.mean(x)

T = 1 / fs_target

X = fft(x)

f = fftfreq(N, T)  # frequencies

fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.lineplot(ax=axes[0], x=f[:N//2], y=np.abs(X[:N//2]), linewidth=(0.8))

axes[0].set_title("ECG_motion frequency spectrum 0-100HZ (fft)")
axes[0].set_xlabel("Frequency (HZ)")

sns.lineplot(ax=axes[1], x=f[:N//6], y=np.abs(X[:N//6]), linewidth=(0.8)) # zoom for better analysis

axes[1].set_title("ECG_motion frequency spectrum 0-30HZ (fft)")
axes[1].set_xlabel("Frequency (HZ)")
```

In the motion ecg signal the heartrate still clearly remains as the most dominant frequency, but the following peaks caused by it are obstructed by noise caused by motion during measuring. (the heart-rate mean has not shifted and the dominant range of frequencies also remains the same (around 0-30HZ).

Next the same for Photoplethysmogram (ppg, as they share the same wavelength I will just take 1 channel from each)

```python
x = ppg_2_red_200_clean
N = len(x)

x = x - np.mean(x)

T = 1 / fs_target

X = fft(x)

f = fftfreq(N, T)  # frequencies

fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.lineplot(ax=axes[0], x=f[:N//2], y=np.abs(X[:N//2]), linewidth=(0.8), color='red')

axes[0].set_title("PPG_1_RED clean frequency spectrum 0-100HZ (fft)")
axes[0].set_xlabel("Frequency (HZ)")

sns.lineplot(ax=axes[1], x=f[:N//24], y=np.abs(X[:N//24]), linewidth=(0.8), color='red') # zoom for better analysis

axes[1].set_title("PPG_1_RED clean frequency spectrum 0-8HZ (fft)")
axes[1].set_xlabel("Frequency (HZ)")
```

ppg in the clean dataset still has the most dominant frequency at the heart rate / pulse frequency which was already established in the prior ecg fft

```python
x = ppg_2_red_200_motion
N = len(x)

x = x - np.mean(x)

T = 1 / fs_target

X = fft(x)

f = fftfreq(N, T)  # frequencies

fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.lineplot(ax=axes[0], x=f[:N//2], y=np.abs(X[:N//2]), linewidth=(0.8), color='red')

axes[0].set_title("PPG_1_RED motion frequency spectrum 0-100HZ (fft)")
axes[0].set_xlabel("Frequency (HZ)")

sns.lineplot(ax=axes[1], x=f[:N//24], y=np.abs(X[:N//24]), linewidth=(0.8), color='red') # zoom for better analysis

axes[1].set_title("PPG_1_RED motion frequency spectrum 0-8HZ (fft)")
axes[1].set_xlabel("Frequency (HZ)")
```

# Task 4

Filter the signal using butterworth and moving average filters using reasonable frequency-
band/smoothing so that you are not filtering out the information carrying signal i.e. filter out
only out-of-band noise.

I will just do the filtering on the examples I did for task 3 

```python
from scipy.signal import butter
from scipy.signal import sosfiltfilt
```

```python
def butter_filter(s, fs, n, f_cut, type='band'):

    sos = butter(n, f_cut, type, output='sos', fs=fs)

    s_filt = sosfiltfilt(sos, s)

    return s_filt
```

```python
f_cut_ecg=[0.5, 35.0]
f_cut_ppg=[0.5, 7.0]

filtered_ecg_clean = butter_filter(ecg_200_clean, fs_target, 4, f_cut_ecg)
filtered_ecg_motion = butter_filter(ecg_200_motion, fs_target, 4, f_cut_ecg)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(12,6)) # for motion and clean

fig.subplots_adjust(hspace=1.2)

i0, i1 = 4000, 6000

filtered_clean_ds = filtered_ecg_clean[i0:i1]
filtered_motion_ds = filtered_ecg_motion[i0:i1]

sns.lineplot(
    ax=axes[0],
    x=t_200[i0:i1],
    y=ecg_200_clean[i0:i1],
    color='darkblue',
    linewidth=1
)
axes[0].set_title("Clean ECG before filtering")
axes[0].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[1],
    x=t_200[i0:i1],
    y=filtered_clean_ds,
    color='darkblue',
    linewidth=1
)
axes[1].set_title("Clean ECG after filtering")
axes[1].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[2],
    x=t_200[i0:i1],
    y=ecg_200_motion[i0:i1],
    color='darkblue',
    linewidth=1
)
axes[2].set_title("Clean ECG before filtering")
axes[2].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[3],
    x=t_200[i0:i1],
    y=filtered_motion_ds,
    color='pink',
    linewidth=1
)
axes[3].set_title("Motion ECG after filtering")
axes[3].set_xlabel("Time (s)")
```

From the previous task it was clear that the important information within ecg exists between the frequencies of 0.5 to 35/40HZ, so i chose that as the cutoff for my filter.

next for ppg

```python
filtered_ppg_2_red_clean = butter_filter(ppg_2_red_200_clean, fs_target, 4, f_cut_ppg)
filtered_ppg_2_red_motion = butter_filter(ppg_2_red_200_motion, fs_target, 4, f_cut_ppg)
```

```python
fig, axes = plt.subplots(4, 1, figsize=(12,6)) # for motion and clean

fig.subplots_adjust(hspace=1.2)

i0, i1 = 4000, 8000

filtered_clean_ds = filtered_ppg_2_red_clean[i0:i1]
filtered_motion_ds = filtered_ppg_2_red_motion[i0:i1]



sns.lineplot(
    ax=axes[0],
    x=t_200[i0:i1],
    y=ppg_2_red_200_clean[i0:i1],
    color='red',
    linewidth=1
)
axes[0].set_title("Clean PPG_2_RED before filtering")
axes[0].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[1],
    x=t_200[i0:i1],
    y=filtered_clean_ds,
    color='red',
    linewidth=1
)


axes[1].set_title("Clean PPG_2_RED filtered")
axes[1].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[2],
    x=t_200[i0:i1],
    y=ppg_2_red_200_motion[i0:i1],
    color='darkred',
    linewidth=1
)

axes[2].set_title("Motion PPG_2_RED before filtering")
axes[2].set_xlabel("Time (s)")

sns.lineplot(
    ax=axes[3],
    x=t_200[i0:i1],
    y=filtered_motion_ds,
    color='darkred',
    linewidth=1
)
axes[3].set_title("Motion PPG_2_RED filtered")
axes[3].set_xlabel("Time (s)")
```

The biggest change of filtering can be seen in the ppg signal of the motion dataset.

Using fft to view the frequency domain for ppg motion already showed that there was a dominant frequency at around 0hz, which the filter removed after setting f_cut to [0.5, 7.0]

ppg_2_red motion after filtering in frequecy domain: (important: missing the big spike at 0hz so filtering seems to have worked)

```python
x = filtered_ppg_2_red_motion
N = len(x)

x = x - np.mean(x)

T = 1 / fs_target

X = fft(x)

f = fftfreq(N, T)  # frequencies

fig, axes = plt.subplots(1, 2, figsize=(12,4))

sns.lineplot(ax=axes[0], x=f[:N//2], y=np.abs(X[:N//2]), linewidth=(0.8), color='red')

axes[0].set_title("PPG_1_RED motion frequency spectrum 0-100HZ (fft) filtered")
axes[0].set_xlabel("Frequency (HZ)")

sns.lineplot(ax=axes[1], x=f[:N//24], y=np.abs(X[:N//24]), linewidth=(0.8), color='red') # zoom for better analysis

axes[1].set_title("PPG_1_RED motion frequency spectrum 0-8HZ (fft) filtered")
axes[1].set_xlabel("Frequency (HZ)")
```

bonus note: I also filtered from 1.0 : 7.0 and the frequency graph became even cleaner but not sure whether that is bad practice as HR can sometimes be lower than 1.0HZ resting
