```python
from scipy.signal import butter
from scipy.signal import sosfiltfilt
from scipy.signal import savgol_filter
from scipy.signal import butter, filtfilt
from scipy.signal import lfilter
from scipy.signal import square
```

```python
f_cut_ecg=[5.0, 15.0] # QRS primarly within this freq range | will attenuate t-wave that is 0 - 5 HZ


# highpass & lowpass --> bandpass
def bandpass_filter(s, fs, fcut, order=4):
    nyq = fs / 2
    b, a = butter(order, [fcut[0]/nyq, fcut[1]/nyq], btype='band')
    return filtfilt(b, a, s)



def derivative_filter(ecg, fs):
    T = 1 / fs

    # 1D five-point stencil
    b = np.array([-1, -2, 0, 2, 1]) / (8 * T)
    
    print(b)
    return filtfilt(b, [1], ecg)

# https://medium.com/@cosmicwanderer/pan-tompkins-algorithm-for-detecting-qrs-waves-29c5f2927906

def moving_average(ECG, N=30):
    
    window  = np.ones((1,N))/N
    ECG_ma  = np.convolve(np.squeeze(ECG),np.squeeze(window), mode='same')
    return ECG_ma
```

```python
ecg_bp_clean = bandpass_filter(clean_ecg['ecg'], 200, f_cut_ecg)
ecg_deriv_clean = derivative_filter(ecg_bp_clean, 200)

ecg_bp_motion = bandpass_filter(motion_ecg['ecg'], 200, f_cut_ecg)
ecg_deriv_motion = derivative_filter(ecg_bp_motion, 200)
```

```python
ecg_squared_clean = ecg_deriv_clean ** 2
ecg_squared_motion = ecg_deriv_motion ** 2

ecg_mwi_clean = moving_average(ecg_squared_clean)
ecg_mwi_motion = moving_average(ecg_squared_motion)
```

# Task 2

peak detection algorithm

```python
from scipy.signal import find_peaks
```

```python
ecg_peaks_clean, ecg_properties_clean= find_peaks(
    ecg_mwi_clean,
    height=np.mean(ecg_mwi_clean),
    distance=int(0.3 * fs),
    prominence=0.2 * np.std(ecg_mwi_clean)
)

ecg_peaks_motion, ecg_properties_motion= find_peaks(
    ecg_mwi_motion,
    height=np.mean(ecg_mwi_motion),
    distance=int(0.2 * fs),
    prominence=0.2 * np.std(ecg_mwi_motion)
)
```

```python
def map_peaks_to_ecg(mwi_peaks, s, fs, search_ms=150):
    mapped_peaks = []
    half_window = int((search_ms / 1000) * fs)
    print(half_window)

    for p in mwi_peaks:
        start = max(p - half_window, 0)
        end   = min(p + half_window, len(s))
        local_max = np.argmax(s[start:end]) + start

        mapped_peaks.append(local_max)

    return np.array(mapped_peaks)
```

```python
clean_ecg_np  = clean_ecg["ecg"].to_numpy()
motion_ecg_np = motion_ecg["ecg"].to_numpy()
```

```python
from scipy.signal import medfilt
```

```python
ppg_deriv_clean = np.diff(clean_ppg['ppg_1_red'])

# Smooth derivative (optional)
ppg_deriv_clean = medfilt(ppg_deriv_clean, kernel_size=5)

# Detect peaks in derivative
ppg_foot_peaks, _ = find_peaks(
    ppg_deriv_clean,
    height=np.percentile(ppg_deriv_clean, 75),
    distance=int(0.4 * fs)
)
```

```python
delays = []

for r in valid_peaks_clean:
    
    ppg_candidates = ppg_foot_peaks[ppg_foot_peaks > r]
    
    if len(ppg_candidates) == 0:
        continue
    
    ppg_foot = ppg_candidates[0]
    delay = (ppg_foot - r) / fs  # seconds
    delays.append(delay)

delays = np.array(delays)
```

```python
MIN_DELAY = 0.05
MAX_DELAY = 0.5

valid_delay_mask = (delays >= MIN_DELAY) & (delays <= MAX_DELAY)
delays_clean = delays[valid_delay_mask]
```

```python
mean_delay = np.mean(delays_clean)
std_delay = np.std(delays_clean)

print(f"Avg ECGPPG delay: {mean_delay*1000:.1f} ms")
print(f"Std deviation: {std_delay*1000:.1f} ms")
```
